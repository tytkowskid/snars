```python
import igraph as ig
```

```python
import matplotlib.pyplot as plt
```

```python
g = ig.Graph.Erdos_Renyi(n=1000, m = 2000)
```

```python
degrees = g.degree()
```

```python
top_nodes = sorted(range(len(degrees)), key = lambda x: degrees[x], reverse = True)[:10]
labels = [str(i) if i in top_nodes else "" for i in range(g.vcount())]
```

```python
layout = g.layout("fr")
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ig.plot(
    g,
    layout = layout,
    vertex_labels = labels,
    bbox = (800, 800),
    margin = 20
)
plt.show()
```

```python
from igraph import Graph
from pyvis.network import Network

# --- Example graph (replace with your own)
n = 500  # number of nodes
p = 0.02  # probability of edge creation
g = Graph.Erdos_Renyi(n=n, p=p)

# Add labels (optional)
g.vs["label"] = [f"Node {i}" for i in range(n)]

# --- Create PyVis Network ---
net = Network(
    notebook=True,        # show in Jupyter notebook
    height="800px",
    width="100%",
    bgcolor="#222222",
    font_color="white"
)

# Add all nodes
for v in g.vs:
    net.add_node(
        v.index,
        label=v["label"],
        title=f"Vertex {v.index}",
        color="#97C2FC"
    )

# Add all edges
for e in g.es:
    src, dst = e.tuple
    net.add_edge(src, dst)

# Optional: enable physics and interaction tools
net.show_buttons(filter_=["physics"])
net.force_atlas_2based()

# Render and show in notebook or browser
net.show("igraph_pyvis_visualization.html")
```
